```python
# imports
import pandas as pd
import numpy as np
import itertools
import glob
import shutil
import os
```

```python
# generating individual folders

val_df = pd.read_csv("./datasets/food/annot/val_info.csv", header=None)
classes = pd.read_csv("./datasets/food/annot/class_list.txt", header=None)
class_ord_names = []
for i in range(len(val_df[0])):
    class_ord_names.append(classes[0][val_df[1][i]].split()[1])
val_df[2] = class_ord_names

for class_index in range(len(classes[0])):
    style_class = classes[0][class_index].split()[1]
    print("Style selected: ", style_class)

    src_dir = "./datasets/food/trainA/"
    dst_dir = "./datasets/food/"+str(style_class)+"/"
    if not os.path.exists(dst_dir):
        os.makedirs(dst_dir)
    for jpgfile in glob.iglob(os.path.join(src_dir, "*.jpg")):
        if jpgfile.split("\\")[1] in list(val_df[val_df[2]==style_class][0]):
            shutil.copy(jpgfile, dst_dir)
```

```python
# pull sample styles
style_mask_paths = []
val_df = pd.read_csv("./datasets/food/annot/val_info.csv", header=None)
classes = pd.read_csv("./datasets/food/annot/class_list.txt", header=None)
class_ord_names = []
for i in range(len(val_df[0])):
    class_ord_names.append(classes[0][val_df[1][i]].split()[1])
val_df[2] = class_ord_names

for class_index in range(len(classes[0])):
    style_class = classes[0][class_index].split()[1]
    print("Style selected: ", style_class)

    filtered_df = val_df[val_df[2] == style_class]
    style_mask_paths.append(filtered_df.sample(1)[0])
```

```python
from faststyletransfer_train import FastStyleTransfer
for index in range(len(style_mask_paths)):
    iters=200
    pairings = [[], []]
    FastStyleTransfer(
        save_model_path="./models/"+classes[0][index].split()[1]+"_style__"+str(iters)+"_iter__vgg19_weights.pth", 
        iterations=iters, 
        style_img="./datasets/food/trainA/"+str(style_mask_paths[index]).split()[1]
        )
    pairings[0].append(classes[0][index].split()[1]); pairings[1].append("./models/"+classes[0][index].split()[1]+"_style__"+str(iters)+"_iter__vgg19_weights.pth")
```
